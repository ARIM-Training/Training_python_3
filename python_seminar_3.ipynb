# 2. 探索的データ分析（データの俯瞰）

## 探索的データ分析（EDA） - データの俯瞰

探索的データ分析（EDA）は、データのパターン、特徴、関係性を把握し、洞察を得るためにデータを調査し要約するプロセスです。EDAは、データ分析や機械学習プロジェクトにおいて重要なステップです。

### EDAの目的

EDAの主な目的は次の通りです：

1. **データの理解**: EDAは、データセットや変数について理解を深めることを支援します。データの構造、内容、品質についての洞察を得ることができます。

2. **データのパターンと分布の特定**: EDAを使用してデータを探索し、パターン、トレンド、分布を特定します。可視化や統計的な手法を用いて、データが異なる変数間でどのように分布しているかを理解します。

3. **異常値と外れ値の検出**: EDAは、データセット内の異常値、外れ値、欠損値を特定するのに役立ちます。これにより、データ品質を確保し、データの前処理やクレンジングについての意思決定を行うことができます。

4. **関係と依存関係の探索**: EDAにより、変数間の関係や依存関係を調査することができます。相関分析、クロス集計、可視化を用いて、データ内の関連や依存関係を特定します。

5. **特徴量の選択とエンジニアリング**: EDAは、特徴量の選択とエンジニアリングのプロセスを導く洞察を提供します。特徴量間の関係を理解することで、モデリングに最も関連性の高い変数を特定することができます。

また、腐食量の時間な変化は次のようにして描くことができます。ここでは、可視化のグラフを事前に関数化させておいています。

```python
def elementplot(Naha,Miyakojima,Fukuoka,Shizuoka,Tsukuba,Choshi,element="element",title="title",loc="lower right"):
    plt.figure(figsize=(12,8))
    plt.style.use('ggplot')
    
    plt.plot(Naha.Date,Naha[element],'-o',label='Nishihara')
    plt.plot(Miyakojima.Date,Miyakojima[element],'-o',label='Miyakojima')
    plt.plot(Fukuoka.Date,Fukuoka[element],'-o',label='Fukuoka')
    plt.plot(Shizuoka.Date,Shizuoka[element],'-o',label='Shimizu')
    plt.plot(Tsukuba.Date,Tsukuba[element],'-o',label='Tsukuba')
    plt.plot(Choshi.Date,Choshi[element],'-o',label='Choshi')
    
    plt.legend(loc=loc,frameon=True,fontsize=12)
    #plt.ylim(0,3500)
    plt.ylabel("{}".format(title),fontsize=20,labelpad=10)
    plt.xlabel("",fontsize=20,labelpad=10)
    plt.tick_params(labelsize=16)
    plt.grid
    plt.show()
```

```python
elementplot(Naha,Miyakojima,Fukuoka,Shizuoka,Tsukuba,Choshi,element="Corrosion",title="Corrosion Rate (g/m2/y)")
```

# 4. 機械学習モデリング

## 0.　準備　トレーニングデータとテストデータの分割（クロスバリデーション）

```python
dfRpf
```

### データセットの分割（Data set splitting）

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.preprocessing import label_binarize

# 説明変数
#['Corrosion', 'NaClmday', 'SO2mday', 'AT', 'HT', 'LT', 'Rain', 'Sun','Power', 'AW', 'MW', 'PMW', 'Hum', 'Lhum']

X = dfR.drop(['Corrosion'], axis=1)
y = dfR.Corrosion

#arrayに変換
X=X.values
y=y.values
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,
                                               random_state=42,
                                               test_size=0.1
                                              )
```

```python
print(X_train.shape,y_train.shape,X_test.shape,y_test.shape)
```

```python
#特徴量のリスト
X_name=list(dfR.drop(['Corrosion'],axis=1).columns)
print(X_name)
```

## 1. LinearRegression
### 機械学習モデルの構築

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)
```

```python
print ("LinearRegression Training Score : {}".format(model.score(X_train, y_train)))
print ("LinearRegression Test Score : {}".format(model.score(X_test, y_test)))
```

訓練データのR2は87%、テストデータでは82%であり、線形回帰そのものでも比較的に良好なモデル構築ができています。
Observed-Calc図で確認をしてみましょう。

### 検証
ここでは、各種のモデル検証で行う一連のツールをRegMLModuleとしてまとめてあります。それをamlとして呼び出して検証を見てみることします。

線形回帰の場合、重要な特徴量としてはHT（最高気温）やHAT(日最高気温の平均)が寄与しているとあります。

## 2. ランダムフォレスト

```python
rf_reg=aml.gridreg.grid_randomforest(X_train,y_train,X_test,y_test,X_name)
```

ランダムフォレストでも訓練データのR2は96%、テストデータでは88%であり、ややOverFitting傾向はみられますが最初のアタリとしては悪くありません。

また、重要な特徴量としてはPMW（最大瞬間風速）やAW（平均風速）、MW(最大風速)が寄与しているとあります。

## 3. サポートベクターマシーン

```python
svm_reg=aml.gridreg.grid_svm(X_train,y_train,X_test,y_test,X_name)
```

サポートベクターでは訓練データでもR2は28%で、テストデータでは-6%となり、モデル構築が上手くできていません。

## 4. 勾配ブースティング

```python
gb_reg=aml.gridreg.grid_gradientboosting(X_train,y_train,X_test,y_test,X_name)
```

勾配ブースティングで訓練データのR2で100%、テストデータでは92%で、こちらもややOverFitting傾向はみられますが最初のアタリとしては悪くありません。

また、重要な特徴量としてはPMW（最大瞬間風速）やAW（平均風速）はランダムフォレストと同じですが、LHum (最小相対湿度)が寄与しているとあります。
