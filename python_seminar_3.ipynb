【腐食観測条件】
1. Corrosion: 腐食量
1. NaClmday: 飛来塩分量
1. SO2mday：亜硫酸ガス量
1. Place：　実験場所


【気象変数】  
気象データは気象庁の観測値であり、次の33変数になります。

1. Date：年月
1. AT :平均気温 [℃]
1. HAT :日最高気温の平均  [℃]
1. LAT :日最低気温の平均  [℃]
1. HT :最高気温 [℃]
1. LT :最低気温 [℃]
1. CAT :日平均気温25℃以上日数　(日)
1. CAT0 :日平均気温0℃未満日数　(日)
1. CHT :日最高気温25℃以上日数　(日)
1. CHT0 :日最高気温0℃未満日数　(日)
1. CLT :日最低気温25℃以上日数　(日)
1. CLT0 :日最低気温0℃未満日数　(日)
1. Rain :降水量の合計　(mm)
1. MRain :日降水量の最大　(mm)
1. Sun:日照時間　(時間)
1. RateSun :日照率　(％)
1. CSun :日照時間0.1時間未満日数　(日)
1. CSun40 :日照率40%以上日数　(日)
1. Power :平均全天日射量　(MJ/㎡)
1. MSnow :最深積雪　(cm)
1. Snow :降雪量合計　(cm)
1. MdaySnow :降雪量日合計最大　(cm)
1. AW :平均風速　(m/s)
1. MW :最大風速　(m/s)
1. MWD :最大風速（風向）(m/s)
1. CMW :日最大風速10m/s以上日数　(日)
1. PMW :最大瞬間風速　(m/s)
1. PMWD :最大瞬間風速（風向）(m/s)
1. WD :最多風向　(16方位)
1. Vap :平均蒸気圧　(hPa)
1. Hum :平均湿度　(％)
1. LHum :最小相対湿度　(％)
1. CMSnow:日最深積雪0cm以上日数　(日)

# 3.グラフの一例

また、腐食量の時間な変化は次のようにして描くことができます。ここでは、可視化のグラフを事前に関数化させておいています。

```python
def elementplot(Naha,Miyakojima,Fukuoka,Shizuoka,Tsukuba,Choshi,element="element",title="title",loc="lower right"):
    plt.figure(figsize=(12,8))
    plt.style.use('ggplot')
    
    plt.plot(Naha.Date,Naha[element],'-o',label='Nishihara')
    plt.plot(Miyakojima.Date,Miyakojima[element],'-o',label='Miyakojima')
    plt.plot(Fukuoka.Date,Fukuoka[element],'-o',label='Fukuoka')
    plt.plot(Shizuoka.Date,Shizuoka[element],'-o',label='Shimizu')
    plt.plot(Tsukuba.Date,Tsukuba[element],'-o',label='Tsukuba')
    plt.plot(Choshi.Date,Choshi[element],'-o',label='Choshi')
    
    plt.legend(loc=loc,frameon=True,fontsize=12)
    #plt.ylim(0,3500)
    plt.ylabel("{}".format(title),fontsize=20,labelpad=10)
    plt.xlabel("",fontsize=20,labelpad=10)
    plt.tick_params(labelsize=16)
    plt.grid
    plt.show()
```

```python
elementplot(Naha,Miyakojima,Fukuoka,Shizuoka,Tsukuba,Choshi,element="Corrosion",title="Corrosion Rate (g/m2/y)")
```

# 4. 機械学習モデリング

### 0.トレーニングデータとテストデータの分割（クロスバリデーション）

```python
dfRpf
```

### データセットの分割（Data set splitting）

```python
from sklearn.model_selection import train_test_split

X = dfRpf.drop(['Corrosion'], axis=1)
y = dfRpf.Corrosion

#arrayに変換
X=X.values
y=y.values
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,
                                               random_state=42,
                                               test_size=0.1
                                              )
```

```python
print(X_train.shape,y_train.shape,X_test.shape,y_test.shape)
```

```python
#特徴量のリスト
X_name=list(dfR.drop(['Corrosion'],axis=1).columns)
print(X_name)
```

## 1. LinearRegression
### 機械学習モデルの構築

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)
```

```python
print ("LinearRegression Training Score : {}".format(model.score(X_train, y_train)))
print ("LinearRegression Test Score : {}".format(model.score(X_test, y_test)))
```

訓練データのR2は0.82、テストデータでは0.80であり、線形回帰そのものでも比較的に良好なモデル構築ができています。Observed-Calc図で確認をしてみましょう。

### 検証
ここでは、各種のモデル検証で行う一連のツールを`RegMLModule`としてまとめてあります。それを`aml`として呼び出して検証を見てみることします。

```python
import RegMLModule as aml
```

```python
line_reg=aml.gridreg.grid_liner(X_train, y_train, X_test, y_test,  X_name)
```

線形回帰の場合、重要な特徴量としてはHT（最高気温）やHAT(日最高気温の平均)が寄与しているとあります。

## 2. ランダムフォレスト

```python
rf_reg=aml.gridreg.grid_randomforest(X_train,y_train,X_test,y_test,X_name)
```

ランダムフォレストでも訓練データのR2は0.95、テストデータでは0.84であり、ややOverFitting傾向はみられますが最初のアタリとしては悪くありません。

また、重要な特徴量としてはPMW（最大瞬間風速）やAW（平均風速）、MW(最大風速)が寄与しているとあります。

## 3. サポートベクターマシーン

```python
svm_reg=aml.gridreg.grid_svm(X_train,y_train,X_test,y_test,X_name)
```

サポートベクターでは訓練データでもR2は0.28で、テストデータでは-0.03となり、モデル構築が上手くできていません。

## 4. 勾配ブースティング

```python
gb_reg=aml.gridreg.grid_gradientboosting(X_train,y_train,X_test,y_test,X_name)
```

勾配ブースティングで訓練データのR2で0.99、テストデータでは0.92、こちらもややOverFitting傾向はみられますが最初のアタリとしては悪くありません。

また、重要な特徴量としてはPMW（最大瞬間風速）やAW（平均風速）はランダムフォレストと同じですが、LHum (最小相対湿度)が寄与しているとあります。
